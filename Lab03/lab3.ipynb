```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Initialize lists to store the metrics
steps = []
total_loss = []
cos = []
mse = []
bce = []
dice = []
ssim = []
kl = []

# Read the log file and extract the metrics
train_log_file = 'train_losses_log.log'

with open(train_log_file, 'r') as f:
    for line in f:
        # Check if the line starts with "epoch" to parse
        if line.startswith("epoch"):
            # Split the line to get epoch and metrics part
            parts = line.split(":", 1)
            step = int(parts[0].split()[3])  # Extract the epoch number
            metrics = ast.literal_eval(parts[1].strip())  # Convert string dict to actual dict
            
            # Append the extracted values to the lists
            steps.append(step)
            total_loss.append(metrics['total_loss'])
            cos.append(metrics['cos'])
            mse.append(metrics['mse'])
            bce.append(metrics['bce'])
            dice.append(metrics['dice'])
            ssim.append(metrics['ssim'])
            kl.append(metrics['kl'])

# Convert lists to DataFrame for easier plotting
df = pd.DataFrame({
    'step': steps,
    'total_loss': total_loss,
    'cos': cos,
    'mse': mse,
    'bce': bce,
    'dice': dice,
    'ssim': ssim,
    'kl': kl
})

# Plotting all metrics in the same figure
plt.figure(figsize=(12, 8))

# Plot each metric on the same graph
plt.plot(df['step'], df['total_loss'], label='Total Loss', color='blue', linewidth=1)
plt.plot(df['step'], df['cos'], label='Cosine Similarity', color='orange', linewidth=1)
plt.plot(df['step'], df['mse'], label='Mean Squared Error', color='green', linewidth=1)
plt.plot(df['step'], df['bce'], label='Binary Cross Entropy', color='red', linewidth=1)
plt.plot(df['step'], df['dice'], label='Dice Coefficient', color='purple', linewidth=1)
plt.plot(df['step'], df['ssim'], label='Structural Similarity', color='brown', linewidth=1)
plt.plot(df['step'], df['kl'], label='Kullback-Leibler Divergence', color='pink', linewidth=1)

# Adding labels and legend
plt.title('Training Loss Metrics')
plt.xlabel('Step')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Initialize lists to store the metrics for the test log
test_steps = []
test_total_loss = []
test_mse = []
test_bce = []
test_dice = []
test_ssim = []

# Read the test log file and extract the metrics
test_log_file = 'test_losses_log.log'

with open(test_log_file, 'r') as f:
    for line in f:
        if line.startswith("epoch"):
            parts = line.split(":", 1)
            step = int(parts[0].split()[3])
            metrics = ast.literal_eval(parts[1].strip())
            
            test_steps.append(step)
            test_total_loss.append(metrics['total_loss'])
            test_mse.append(metrics['mse'])
            test_bce.append(metrics['bce'])
            test_dice.append(metrics['dice'])
            test_ssim.append(metrics['ssim'])

# Convert lists to DataFrame for easier plotting
test_df = pd.DataFrame({
    'step': test_steps,
    'total_loss': test_total_loss,
    'mse': test_mse,
    'bce': test_bce,
    'dice': test_dice,
    'ssim': test_ssim,
})

# Plotting all metrics in the same figure for test log
plt.figure(figsize=(12, 8))

# Plot each metric on the same graph for test log
plt.plot(test_df['step'], test_df['total_loss'], label='Total Loss', color='blue', linewidth=2)
plt.plot(test_df['step'], test_df['mse'], label='Mean Squared Error', color='green', linewidth=2)
plt.plot(test_df['step'], test_df['bce'], label='Binary Cross Entropy', color='red', linewidth=2)
plt.plot(test_df['step'], test_df['dice'], label='Dice Coefficient', color='purple', linewidth=2)
plt.plot(test_df['step'], test_df['ssim'], label='Structural Similarity', color='brown', linewidth=2)

# Adding labels and legend
plt.title('Test Loss Metrics')
plt.xlabel('Step')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
metrics_steps = []
metrics_iou = []
metrics_mse = []
metrics_ssim = []

# Read the metrics log file and extract the metrics
metrics_log_file = 'metrics_log.log'

with open(metrics_log_file, 'r') as f:
    for line in f:
        if line.startswith("epoch"):  # Assuming the file starts with "step" keyword
            parts = line.split(":", 1)
            step = int(parts[0].split()[3])
            metrics = ast.literal_eval(parts[1].strip())
            
            metrics_steps.append(step)
            metrics_iou.append(metrics['IoU'])
            metrics_mse.append(metrics['mse'])
            metrics_ssim.append(metrics['ssim'])

# Convert lists to DataFrame for easier plotting
metrics_df = pd.DataFrame({
    'step': metrics_steps,
    'iou': metrics_iou,
    'mse': metrics_mse,
    'ssim': metrics_ssim,
})

# Plotting all metrics in the same figure for metrics log
plt.figure(figsize=(12, 8))

# Plot each metric on the same graph for metrics log
plt.plot(metrics_df['step'], metrics_df['iou'], label='IoU', color='blue', linewidth=2)
plt.plot(metrics_df['step'], metrics_df['mse'], label='Mean Squared Error', color='orange', linewidth=2)
plt.plot(metrics_df['step'], metrics_df['ssim'], label='Structural Similarity', color='green', linewidth=2)

# Adding labels and legend
plt.title('Metrics Log: IoU, MSE and SSIM')
plt.xlabel('Steps')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
